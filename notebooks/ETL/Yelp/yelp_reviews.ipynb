Cargando los Datasets a Dataframes

```python
yelp_reviews = pd.read_ndjson("datasets/yelp/review.json", lines=True)

# Utilizando el Archivo que ya se habia pasado a parquet
yelp_business = pd.read_parquet("datasets/yelp/yelp_business.parquet")
```

Observando como viene los Negocios

```python
print(yelp_business.shape)
yelp_business.head()
```

Observando cuantos estados "No Disponible" hay

```python
mask = yelp_business['state'] == 'No Disponible'

yelp_business.loc[mask].shape[0]
```

Rellenando manualmente los únicos 3 estados faltantes

```python
"""
estados_faltantes
    "latitude": 34.426679,
    "longitude": -119.711197,
    "Estado": "CA"

    "latitude": 38.551126,
    "longitude": -90.335695,
    "Estado": "MO"

    "latitude": 32.223236,
    "longitude": -110.880452,
    "Estado": "AZ"
"""

yelp_business.iloc[0,4] = "CA"

yelp_business.iloc[1,4] = "MO"

yelp_business.iloc[2,4] = "AZ"

#Comprobando
mask = yelp_business['state'] == 'No Disponible'

print(yelp_business.loc[mask].shape[0])
yelp_business.head()
```

Obteniendo en una lista todos los id de los negocios de los 5 estados a los cuales se les hará el estudio

```python
"""
Ney Yok       ---> NY
Florida       ---> FL
Texas         ---> TX
Nuevo Mexico  ---> NM
Pennsylvania  ---> PA
"""
lista_estados = ["NY", "FL", "TX", "NM", "PA"]

mask = yelp_business['state'].isin(lista_estados)

business_Top_5_estados = yelp_business.loc[mask]

lista_id_Top_5_estados = business_Top_5_estados['business_id'].to_list()


print(f'Total Negocios en Top 5 Estados {len(lista_id_Top_5_estados)}')
```

Filtrando los reviews de los 5 Estados

```python
mask = yelp_reviews['business_id'].isin(lista_id_Top_5_estados)

yelp_reviews_top_5_estados = yelp_reviews.loc[mask]

print(f'Total Reviews en Top 5 Estados {yelp_reviews.shape[0]}')
```

Guardando Tablas

```python
# Tabla de reviews completa
yelp_reviews.to_parquet("datasets/yelp/yelp_review.parquet")

# Tabla de reviews solo 5 estados 
yelp_reviews_top_5_estados.to_parquet("datasets/yelp/review_top_5_estados.parquet")

# Negocios solo 5 estados
yelp_reviews_top_5_estados.to_parquet("datasets/yelp/review_top_5_estados.parquet")
```
