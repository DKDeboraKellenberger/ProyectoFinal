```python
import pandas as pd

sitios = pd.read_parquet(r'C:\Users\mauri\OneDrive\Escritorio\proyectogrupal\data\google_sitios.parquet', engine='pyarrow')
sitios.info()
```

```python
address = sitios[['gmap_id','address', 'name']]
address.head()
```

```python
# Separar la columna 'address' en nuevas columnas
address_split = address['address'].str.split(',', n=3, expand=True)

# Cambiar el nombre de las columnas
address_split.columns = ['nombre', 'direccion', 'ciudad', 'cod.postal']
```

```python
address = address[['gmap_id']].join(address_split)
address.head()
```

```python
# Agregar una columna nueva llamada estados.
address['estado'] = None

## me di cuenta que el cod.postal contiene las letras del estado, ahi seria mas facil.
address.loc[address['cod.postal'].str.contains('CA', na=False), 'estado'] = 'California'
address.loc[address['cod.postal'].str.contains('NY', na=False), 'estado'] = 'Nueva York'
address.loc[address['cod.postal'].str.contains('IA', na=False), 'estado'] = 'Iowa'
address.loc[address['cod.postal'].str.contains('GA', na=False), 'estado'] = 'Georgia'
address.loc[address['cod.postal'].str.contains('FL', na=False), 'estado'] = 'Florida'
address.loc[address['cod.postal'].str.contains('TX', na=False), 'estado'] = 'Texas'
address.loc[address['cod.postal'].str.contains('LA', na=False), 'estado'] = 'Louisiana'
address.loc[address['cod.postal'].str.contains('OR', na=False), 'estado'] = 'Oregon'
address.loc[address['cod.postal'].str.contains('WV', na=False), 'estado'] = 'West Virginia'
address.loc[address['cod.postal'].str.contains('VA', na=False), 'estado'] = 'Virginia'
address.loc[address['cod.postal'].str.contains('AR', na=False), 'estado'] = 'Arkansas'
address.loc[address['cod.postal'].str.contains('PA', na=False), 'estado'] = 'Pennsylvania'
address.loc[address['cod.postal'].str.contains('NM', na=False), 'estado'] = 'New Mexico'
address.loc[address['cod.postal'].str.contains('NC', na=False), 'estado'] = 'North Carolina'
address.loc[address['cod.postal'].str.contains('TN', na=False), 'estado'] = 'Tennessee'
address.loc[address['cod.postal'].str.contains('WI', na=False), 'estado'] = 'Wisconsin'
address.loc[address['cod.postal'].str.contains('NJ', na=False), 'estado'] = 'New Jersey'
address.loc[address['cod.postal'].str.contains('IN', na=False), 'estado'] = 'Indiana'
address.loc[address['cod.postal'].str.contains('IL', na=False), 'estado'] = 'Illinois'
address.loc[address['cod.postal'].str.contains('DC', na=False), 'estado'] = 'District of Columbia'
address.loc[address['cod.postal'].str.contains('MD', na=False), 'estado'] = 'Maryland'
address.loc[address['cod.postal'].str.contains('ME', na=False), 'estado'] = 'Maine'
address.loc[address['cod.postal'].str.contains('NE', na=False), 'estado'] = 'Nebraska'
address.loc[address['cod.postal'].str.contains('WA', na=False), 'estado'] = 'Washington'
address.loc[address['cod.postal'].str.contains('MI', na=False), 'estado'] = 'Michigan'
address.loc[address['cod.postal'].str.contains('OH', na=False), 'estado'] = 'Ohio'
address.loc[address['cod.postal'].str.contains('OK', na=False), 'estado'] = 'Oklahoma'
```
